```python
import pandas as pd
```

```python
df=pd.read_table('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv')
```

```python
df.head()
```

```python
#removing dollar and converting into float
```

```python
rem_dollar=lambda x:float(x.split('$')[1])
df['item_price']=df['item_price'].apply(rem_dollar)
```

```python
df.dtypes
```

```python
df['Total_price']=df['quantity']*df['item_price']
```

```python
#Order id that gave max transcation value
```

```python
c=df.groupby('order_id')["Total_price"].sum()
c[c==max(c)].index
```

```python
#Lets check orderid 1443
df[df['order_id']==1443]
```

```python
filtered=df.drop_duplicates(['item_name','quantity'])
one_prod=filtered[filtered['quantity']==1]
price_per_item = one_prod[['item_name', 'item_price']]
price_per_item.sort_values(by = "item_price", ascending = False)
```
